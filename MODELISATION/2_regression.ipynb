### Importation des fichiers et des librairies :

```python
import pandas as pd 
X_train_enc = pd.read_csv('X_train_enc.csv')
X_test_enc = pd.read_csv('X_test_enc.csv')
```

```python
y_train = pd.read_csv('y_train_reg.csv')
y_test = pd.read_csv('y_test_reg.csv')
```

## SGD Regressor (instanciation, entrainement, test et évaluation) :

```python
from sklearn.linear_model import SGDRegressor
sgd_reg = SGDRegressor(max_iter = 1000, eta0 = 0.1, penalty = None, random_state = 42)
sgd_reg.fit(X_train_enc, y_train)
```

```python
pred_sgd = sgd_reg.predict(X_test_enc)
```

```python
print(f"""le score R2 du modèle sgd regressor sur le jeu d'entrainement est de 
    {sgd_reg.score(X_train_enc, y_train)}""")
print(f"""le score R2 du modèle sgd regressor sur le jeu de test est de 
    {sgd_reg.score(X_test_enc, y_test)}""")
```
